**Task 1: Load the Titanic dataset**

```python
# TODO: Load the Titanic dataset from the following URL
import pandas as pd

# I used the file they provided instead of the url

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

**Task 2: Inspect the data**

```python
# TODO: Display the first 5 rows of the dataset
df.head(5)
```

```python
# TODO: Display the last 5 rows of the dataset
df.tail(5)
```

```python
# TODO: Display the data types of each column
df.dtypes
```

```python
# TODO: Display summary statistics of the numerical columns
df.describe()
```

```python
# Task 3: Clean and preprocess the data

# TODO: Create a new categorical feature 'PclassCategory' from 'Pclass'

df['PclassCategory'] = df['Pclass'].map({1: 'Upper', 2: 'Middle', 3: 'Lower'})
df.head()
```

```python
# TODO: Create a new categorical feature 'CabinCategory' from 'Cabin'

# Creating categories based on the first letter, so I'm extracting the first letter & assigning 'Unknown' for NaNs
df['CabinCategory'] =df['Cabin'].apply(
    lambda x: x[0] if pd.notnull(x) else 'Unknown'
)

df['CabinCategory'].unique()
```

```python
# TODO: Handle missing values in the 'Age' column (use mean age)

df['Age'].fillna(df['Age'].mean(), inplace = True)

print("After filling missing values in Age column:\n")
df.isnull().sum()
```

```python
# TODO: Create a 'Family' feature by combining 'SibSp' and 'Parch'

df['Family'] = df['SibSp'] + df['Parch']


print(df[['SibSp', 'Parch', 'Family']].head())
```

**Task 4: Identify specific data points**

```python
# TODO: Find passengers with the highest fare

highest_fare_passengers = df[df['Fare'] == df['Fare'].max()]
highest_fare_passengers
```

```python
# TODO: Find passengers with the largest family size

largest_family_passengers = df[df['Family'] == df['Family'].max()]
largest_family_passengers
```

**Task 5: Demonstrate Pandas Series and DataFrame functionalities**

```python
# TODO: Create a Pandas Series of passenger names

names_series = df['Name']
names_series
```

```python
# TODO: Create a DataFrame of passenger names and ages
names_ages_df = df[['Name', 'Age']]
names_ages_df
```

```python
# TODO: Slice the DataFrame to show only passengers older than 50

older_passengers = names_ages_df[names_ages_df['Age'] > 50]
older_passengers
```

**Task 6: Introduce multi-indexing**

```python
# TODO: Create a multi-index DataFrame grouped by 'Sex' and 'Pclass'

multi_index_df = df.groupby(['Sex', 'Pclass']).mean(numeric_only=True)
multi_index_df
```

**Task 7: Visualizations**

```python
# TODO: Create a histogram of passenger ages
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins = 10, color = 'lightcoral', edgecolor= 'red')
plt.title('Histogram of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Most of the Titanic passengers are at around age 25-32.
# There are few passengers above 55 years
```

```python
# TODO: Create a bar plot of survival rates by passenger class

survival_rates = df.groupby('Pclass')['Survived'].mean()

plt.figure(figsize=(10, 6))
plt.bar(x=survival_rates.index, height= survival_rates.values, color='lightblue')
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate of the 1st class passengers was the highest at 60% & 3rd class was the lowest at 20%
# which makes sense since 1st class passengers are usually taken care of more.
```

```python
# TODO: Create a heatmap of correlation between numerical features

import seaborn as sns

correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap = 'coolwarm' , fmt= ".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

# We have a high positive correlation between: SibSp & Family (0.89) + Parch & Family (0.78) --> As siblings, spouse, parents & children numbers increase so does the Family number.
# We have a moderate negative correlation between: Pclass & Fare, so the higher the Pclass number the lower the Fare --> Pclass 3 had the lowest fare compared to 1
# All the other variables have weak relationships.
```

```python
# TODO: Create a pair plot of 'Age', 'Fare', 'SibSp', and 'Parch', colored by 'Survived'

selected_columns = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']

# Create the pair plot
sns.pairplot(df[selected_columns], hue='Survived', palette='Set1')

# Show the plot
plt.suptitle('Pair Plot of Age, Fare, SibSp, and Parch by Survival', y=1.02)
plt.show()
```

**Task 8: Analyze survival rates**

```python
# TODO: Calculate and display survival rates by sex

survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Survival Rates by Sex:")
survival_by_sex
```

```python
# TODO: Calculate and display survival rates by passenger class

survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("Survival Rates by Passenger Class:")
survival_by_class
```

**Task 9: Analyze age distribution**

```python
# TODO: Create a box plot of age distribution by passenger class and sex

plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Pclass'], y=df['Age'], hue= df['Sex'])
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Sex')
plt.title('Age Distribution by Passenger Class and Sex')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()

# Passenger class 1 had the oldest age group (25 - 50 years) & male's were also the oldest compared to females.
# Passenger class 2 had alot of outliers & the age group was between 22 & 45 Male and Female were equally aged.
# Passenger class 3 had the youngest age group (18 - 30 years), which means it was mainly young people that could afford the cheapest tickets.
```

**Task 10: Analyze fare distribution**

```python
# TODO: Create a violin plot of fare distribution by passenger class and survival status

plt.figure(figsize=(12, 6))
sns.violinplot(x=df['Pclass'], y=df['Fare'], hue=df['Survived'], palette='muted')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.show()

```
