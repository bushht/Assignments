**Task 9: Iterators and Generators**

```python
# Create an iterator for the first 5 even numbers.

# Create an iterator from range(10)
even_iter = iter(range(10))

# Print first 5 even numbers using next()
count = 0
while count < 5:
    num = next(even_iter)
    if num % 2 == 0:
        print(num)
        count += 1

```

```python
# Then, create a generator for the same.

def even_numbers_generator(count):
    num = 0
    for _ in range(count):
        yield num
        num += 2

# Use the generator
for num in even_numbers_generator(5):
    print(num)
```

**Task 10: Map, Reduce, and Filter**

```python
# Use map to square all numbers in a list.

numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x**2, numbers))
print("Squared numbers:", squares)
```

```python
# Use reduce to find the product of all numbers in a list.
# It performs repeated operations on a list, combining all items into a single result by applying a function pairwise.

from functools import reduce

# reduce(function, iterable)

product = reduce(lambda x, y: x * y, numbers)
print("Product of numbers:", product)
```

```python
# Use filter to get only even numbers from a list.
# Used to filter elements from a list (or any iterable) based on a condition.

evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", evens)
```

**Task 11: Object-Oriented Programming - Creating a Class**

```python
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Define the Rectangle class
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create an object of the class
my_rectangle = Rectangle(10, 5)

# Access the attributes
print("Length:", my_rectangle.length)
print("Width:", my_rectangle.width)
print("Area:", my_rectangle.area())
print("Perimeter:", my_rectangle.perimeter())
```
